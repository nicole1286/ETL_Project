```python
# import the library we use to open URLs
import urllib.request
import requests
```

```python
# specify which URL/web page we are scraping
url="https://en.wikipedia.org/wiki/List_of_countries_by_system_of_government"
```

```python
# open the URL using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup as bs
```

```python
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = bs(page, 'lxml')
```

```python
#print(soup.prettify())
```

```python
# use the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store
all_tables=soup.find_all('table')
#all_tables
```

```python
# look thru all_tables for wikitable sortable for table we want
right_table=soup.find('table', class_='wikitable sortable')
#right_table
```

```python
# To look thru the rows, setup array to store strings
A=[]
B=[]
C=[]
D=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==4:
        A.append(cells[0].find('a',text=True).text)
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        
print(A[:5])
```

```python
import pandas as pd
df=pd.DataFrame(A, columns=['Country'])
df['Type']=B
df['Head']=C
df['Basis_Legit']=D
df
```

```python
# write file to csv
df.to_csv(r'governments.csv', index=False)
```
